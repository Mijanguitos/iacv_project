```python
from pathlib import Path
import numpy as np
import cv2 as cv2
import json
import os
```

```python
selected_points = []

def click_event(event, x, y, flags, param):
    """
    Mouse callback function to capture left-clicks.
    """
    if event == cv2.EVENT_LBUTTONDOWN:
        # Print the coordinates to the console
        print(f"Point selected at: (X: {x}, Y: {y})")
        
        # Store the point in our list
        selected_points.append((x, y))
        
        # Draw a small green circle on the image for visual feedback
        cv2.circle(frame, (x, y), 5, (0, 255, 0), -1)
        
        # Optionally, display the coordinates on the image itself
        cv2.putText(frame, f"{x},{y}", (x + 10, y - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)
        
        # Update the window to show the new circle/text
        cv2.imshow(window_name, frame)
```

```python

# --- Main Execution ---


# 1. Define the video path (replace with your actual video file)
PROJECT_ROOT = Path().resolve().parent.parent
clip = "clip_11.mp4"
video_path = f"{PROJECT_ROOT}\\data\\clips\\{clip}"


# 2. Open the video capture object
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Error: Could not open video file '{video_path}'. Please check the path.")
    exit()

# 3. Read the first frame
success, frame = cap.read()

# 4. Release the video object immediately since we only need the first frame
cap.release()

if not success:
    print("Error: Could not read the first frame of the video.")
    exit()

# 5. Create a named window and attach the mouse callback function
window_name = "First Frame - Click to select points (Press 'q' or ESC to exit)"
cv2.namedWindow(window_name)
cv2.setMouseCallback(window_name, click_event)

# 6. Display the initial frame
cv2.imshow(window_name, frame)
print("Click on the image to select points. Press 'q' or the ESC key when you are finished.")

# 7. Keep the window open until the user presses 'q' or 'ESC'
while True:
    # Wait for 1 millisecond for a key press
    key = cv2.waitKey(1) & 0xFF
    
    # Break the loop if 'q' (113) or ESC (27) is pressed
    if key == ord('q') or key == 27:
        break

# 8. Clean up and close all windows
cv2.destroyAllWindows()

# 9. Output the final list of collected points
print("\nFinal list of selected points:")
print(selected_points)
selected_points = []
```
